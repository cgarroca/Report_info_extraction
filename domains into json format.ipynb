```python
import json
#now add the subdomains
#for i in list(range(len(subs[0]))):
#    

#domains[0]['sub_doms']
#doms[0].index('1')
#doms[0].index(subs[0][1])

domains=[]
for i in list(range(len(doms[0]))):
    domains.append({'code':doms[0][i],'desc':doms[1][i],'sub_doms':[]})
    for j in list(range(len(subs[0]))):
        if subs[0][j] == doms[0][i]:
            domains[i]['sub_doms'].append({'code':subs[1][j],'desc':subs[2][j]})
            
for i in list(range(len(doms[0]))):
    for j in list(range(len(domains[i]['sub_doms']))):
        tm=list(set(sub_subs[0]))
        for l in list(range(len(tm))):
            code = domains[i]['sub_doms'][j]['code']
            if tm[l]== code:
                domains[i]['sub_doms'][j]['sub_sub_doms']=[]
                for k in list(range(len(sub_subs[0]))):
                    if sub_subs[0][k]== code:
                        domains[i]['sub_doms'][j]['sub_sub_doms'].append({'code':sub_subs[1][k],'desc':sub_subs[2][k]})
                        
print(json.dumps(domains, indent = 4))
#Classification of International Statistical Activities

with open('Classification Statistical Activities.json', 'w') as json_file:  
    json.dump(domains, json_file,indent = 4)
```
